## SMOTE
#### what smote technique does is it takes two points of the minority target and between the distance of those two points
#### it fills with different values. It is an upsampling technique

```python
from sklearn.datasets import make_classification
```

```python
##weight is for the weight of each class in the category that we want here 0.90 means an imbalanced class
X,y=make_classification(n_samples=1000,n_redundant=0, n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df1,df2],axis=1)
```
